```python
#SHOW THE DATA FRAME Vaccination analyse 2022
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt

drive.mount('/content/drive')

file_path = "/content/drive/My Drive/Vaccination analyse 2022.xlsx"

df = pd.read_excel(file_path)
df2 = df.head(20)
df2.style.hide(axis="index")
```

```python

#THE BAR CHART SHOW WHICH IS THE BEST POPULATED COUNTRY
df2 = df.head(20)

plt.figure(figsize=(10, 6))
plt.bar(df2['Country'], df2['Total Population'], color='green')


plt.xlabel("Name")
plt.ylabel("Value")
plt.title("Bar Chart of Country vs Total Population")

plt.xticks(rotation=45, ha='right')
plt.tight_layout()


plt.show()
```

```python
#Scatter Plot SHOWS ACTIVE CASES
plt.figure(figsize=(22, 8))
plt.scatter(df2['Country'], df2['Active Cases'], color='black', marker='X')
plt.title("Top 20 Data - Scatter Plot")
plt.show()
```

```python
#VACCINATION COUNT THE COUNTRY
df2_filtered = df2[df2['Country'] != 'China']

plt.figure(figsize=(18, 12))
plt.pie(df2['Vaccination Count'], labels=df2['Country'], autopct='%1.1f%%', startangle=140)
plt.title("Top 20 Data - Pie Chart")
plt.show()

```

```python
#THE GRAPH SHOWS TOTAL TESTS OF VACCINATION PEOPLES OF COUNTRY
plt.figure(figsize=(16, 6))
plt.plot(df2['Country'], df2['Total Tests'], marker='o', linestyle=':', color='b')
plt.title("Top 20 Data - Line Chart")
plt.show()
```

```python
#Vaccination Rate
plt.figure(figsize=(14, 6))
plt.bar(df2['Country'], df2['Vaccination Rate (%)'], color='lightblue', alpha=0.7, label="Bar Chart")

plt.plot(df2['Country'], df2['Vaccination Rate (%)'], marker='o', linestyle='-', color='red', label="Line Chart")

plt.xlabel("Country")
plt.ylabel("Vaccination Rate (%)")
plt.title("Bar Chart & Line Chart Combined")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
